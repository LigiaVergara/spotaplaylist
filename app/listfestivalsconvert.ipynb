```python
def transform_festivals(festival_dict):
    transformed_festivals = []
    for festival_name, festival_data in festival_dict.items():
        transformed_festival = {
            "name": festival_data["name"],
            "url": festival_data["url"],
            "image": festival_data["image_url"],
            "country": festival_data["country"],
            "artists": festival_data["artists"]
        }
        transformed_festivals.append(transformed_festival)
    return transformed_festivals

# Transform the input festivals dictionary
transformed_festivals = transform_festivals(input_festivals)

# Output the transformed list
print(transformed_festivals)
```
